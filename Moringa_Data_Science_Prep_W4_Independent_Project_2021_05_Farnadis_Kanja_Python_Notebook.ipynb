```python
# Let's the Pandas Library

import pandas as pd

# Let's the Numpy Library

import numpy as np
```

```python
#because the two columns scheduled at and displayed comment have so many null values, we drop them 
df.drop(["Scheduled at", "Displayed comment"], axis = 1, inplace = True) 
df.head()
```

```python
#drop geo point column as it seems irrelevant to this analysis
df.drop(["Geo point"], axis = 1, inplace = True) 
df.head
```

```python
# We count the number of non - missing values in the df dataframe

df.count()
```

```python
#determine if there are any zero values in the data after some cleaning
np.count_nonzero(df.isnull())
```

```python
#set the all the 5 date and time columns into one
df['datetime'] = pd.to_datetime(df[['minute','hour','month','day','year']])
df.head(5)
```

```python
#delete the older date columns
df.drop(["year", "month", "day", "hour", "minute"], axis = 1, inplace = True) 
df.head(5)
```

```python
#splitting the datetime to two separate columns 'time' and 'date'
df['date']=[d.date() for d in df.datetime]
df['time']=[d.time() for d in df.datetime] 
df['date']=pd.to_datetime(df['date'])
df['time']=pd.to_datetime(df['time'], format="%H:%M:%S:%f", errors='ignore')
```

```python
#check the first few rows
df.head(5)
```

```python
#delete the datetime column
df.drop(["datetime"], axis = 1, inplace = True) 
df.head(5)
```

```python
#checking if the changes have been effected
df.dtypes
```

```python
#changing columns to lower case
#renaming columns
df.columns = map(str.lower, df.columns)
columns=['address', 'cars','bluecar_counter','utilib_counter', 'utilib_1.4_counter', 'charge_slots','charging_status', 'city', 'id', 'kind', 'postal_code',
         'public_name', 'rental_status', 'slots', 'station_type',
         'status', 'subscription_status','date', 'time']
df.columns=columns
df.columns
```

```python
#arranging the time and dates chronologically
df = df.sort_values(by="time" and "date", ascending = True)
df.tail()
```
